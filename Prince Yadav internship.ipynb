# Min_Max_Normalization

```python
cols = ["CRIM", "ZN", "TAX", "B"]

for col in cols:  
    minimum = min(df[col])
    maximum = max(df[col])
    df[col] = (df[col] - minimum) / (maximum - minimum)
```

```python
# Below we can notice that the values of columns CRIM, ZN, TAX, B have been normalized to 0 to 1.
df.hist(figsize = (20, 15)) 
```

# Train_test_split

```python
from sklearn.model_selection import train_test_split

X = df.drop(["MEDV", "RAD"], axis = 1)
y = df["MEDV"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7) 
```

# Model Training (Cross_validation)



```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error 
```

```python
def train(model, X, y):

    model.fit(X_train, y_train)
    
    # predicting on the trainig dataset
    y_pred = model.predict(X_test) 
    
    # Performing Cross_validation
    cv_score = cross_val_score(model, X, y, scoring = "neg_mean_squared_error", cv = 5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("MSE: ", mean_squared_error(y_test, y_pred))
    print("RMSE: ", np.sqrt(mean_squared_error(y_test, y_pred)))
    print("CV score: ", cv_score)
```

# Applying LinearRegression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
train(model, X, y)
y_pred = model.predict(X_test) 
coef = pd.Series(model.coef_, X.columns).sort_values()
coef.plot(kind = "bar", title = "Model Coefficients")

# Here we can see in Linear Regression, ZN has highest positive coefficient and NOX has highest negative coefficient
```

# Applying DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")

# Decision Tree is performing well on MSE and RMSE than LinearRegression but on the CV_score, LinearRegression is higher
# than the Decision Tree
```

# Applying RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")

# As we already know that RM and LSTAT have high correlation with the target variable that is why they both have higher
# feature importance than other attributes. 
# RandomForestRegressor, gives us the best result on MSE and cv_score so far.
```

# Applying ExtraTreeRegressor

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")

# Here we can notice that the model is performing well with ExtraTreesRegressor as MSE, RSMSE AND CV_score have 
# decreased
```

```python
import xgboost as xgb
model = xgb.XGBRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")
```
